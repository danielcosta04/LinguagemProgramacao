# Visualização de Dados para Decisões com Matplotlib e Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

df = pd.read_csv('vendas_brasil_clean_aula5.csv')
display(df.head())
```

```python
df.info()
```

```python
print(f"Tamanho da base de dados: {df.shape[0]} linhas e {df.shape[1]} colunas")
```

## 3. Visualização de Dados

### Receita Total por Estado (`uf`)

Vamos começar visualizando a receita total gerada por cada estado (`uf`). Um gráfico de barras é ideal para comparar os valores entre as diferentes unidades federativas.

```python
# Calcular a receita total por estado
receita_por_uf = df.groupby('uf')['receita'].sum().sort_values(ascending=False).reset_index()

# Criar o gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x='uf', y='receita', data=receita_por_uf, palette='viridis')
plt.title('Receita Total por Estado (UF)')
plt.xlabel('Estado (UF)')
plt.ylabel('Receita Total')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Receita Total por Canal de Venda (`canal_venda`)

Agora, vamos analisar a receita total gerada por cada canal de venda. Isso nos permitirá entender a performance de cada um e identificar os canais mais lucrativos.

```python
# Calcular a receita total por canal de venda
receita_por_canal = df.groupby('canal_venda')['receita'].sum().sort_values(ascending=False).reset_index()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x='canal_venda', y='receita', data=receita_por_canal, palette='magma')
plt.title('Receita Total por Canal de Venda')
plt.xlabel('Canal de Venda')
plt.ylabel('Receita Total')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Receita Total por Segmento de Cliente (`segmento`)

Vamos agora visualizar a receita total gerada por cada segmento de cliente. Isso nos ajudará a comparar o desempenho entre os segmentos 'Varejo' e 'Corporativo'.

```python
# Calcular a receita total por segmento de cliente
receita_por_segmento = df.groupby('segmento')['receita'].sum().sort_values(ascending=False).reset_index()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x='segmento', y='receita', data=receita_por_segmento, palette='plasma')
plt.title('Receita Total por Segmento de Cliente')
plt.xlabel('Segmento de Cliente')
plt.ylabel('Receita Total')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Receita Total por Categoria de Produto (`categoria`)

Vamos agora analisar a receita total gerada por cada categoria de produto. Isso é essencial para entender o desempenho das diferentes categorias e identificar as mais importantes.

```python
# Calcular a receita total por categoria de produto
receita_por_categoria = df.groupby('categoria')['receita'].sum().sort_values(ascending=False).reset_index()

# Criar o gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x='categoria', y='receita', data=receita_por_categoria, palette='rocket')
plt.title('Receita Total por Categoria de Produto')
plt.xlabel('Categoria de Produto')
plt.ylabel('Receita Total')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Análise das Variáveis para Visualização

Com base nas informações do `df.info()` e na estrutura do `df.head()`, podemos inferir o seguinte sobre as variáveis:

### Variáveis para Comparação:
*   **`uf`**: Unidade Federativa (estado) - Permite comparar vendas, categorias ou segmentos entre diferentes estados.
*   **`canal_venda`**: Canal de Venda (Marketplace, Online, Televendas) - Útil para comparar o desempenho de cada canal.
*   **`segmento`**: Segmento de Cliente (Varejo, Corporativo) - Permite comparar o comportamento de compra entre segmentos.
*   **`categoria`**: Categoria do Produto (Acessórios, Móveis, etc.) - Essencial para comparar o desempenho de diferentes categorias.
*   **`mes`**: Mês - Permite comparar o desempenho de vendas ou lucro entre diferentes meses.

### Variáveis para Tendência:
*   **`data`**: Coluna de data - Fundamental para analisar tendências de vendas, lucros ou outras métricas ao longo do tempo (diária, semanal, mensal, anual).

### Variáveis para Distribuição:
*   **`quantidade`**: Quantidade de itens vendidos - A distribuição da quantidade pode revelar padrões de compra (e.g., a maioria dos clientes compra poucos itens ou muitos).
*   **`preco_unitario`**: Preço unitário - A distribuição dos preços pode indicar faixas de preço mais comuns, produtos de alto valor ou promoções.
*   **`receita`**: Receita total - A distribuição da receita por transação pode mostrar o valor médio das vendas e identificar grandes transações.
*   **`lucro`**: Lucro total - A distribuição do lucro pode indicar a rentabilidade das vendas.
*   **`margem_lucro`**: Margem de lucro - A distribuição da margem pode mostrar a saúde financeira das vendas.

### Variáveis para Relação entre Métricas:
*   **`quantidade`**, **`preco_unitario`**, **`receita`**, **`lucro`**, **`margem_lucro`**: Estas variáveis numéricas são cruciais para analisar relações entre si (e.g., como o preço unitário afeta a receita, como a quantidade vendida influencia o lucro).
*   **`categoria` x `receita` / `lucro`**: Para entender quais categorias são as mais rentáveis.
*   **`uf` x `receita` / `lucro`**: Para identificar os estados com maior desempenho financeiro.
*   **`canal_venda` x `segmento` x `receita` / `lucro`**: Para entender o desempenho de cada canal e segmento combinados.

### Receita Total Mensal (`mes`)

Para analisar a tendência da receita ao longo do tempo, vamos visualizar a receita total gerada por cada mês. Um gráfico de linha é o mais adequado para essa análise, pois mostra a progressão contínua.

```python
# Converter a coluna 'data' para datetime, se ainda não for
df['data'] = pd.to_datetime(df['data'])

# Extrair o número do mês (1-12)
df['mes_numero'] = df['data'].dt.month

# Definir a ordem correta dos meses para o plot
ordem_meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

# Mapear o número do mês para o nome do mês em português
# Criar um dicionário de mapeamento: {1: 'Janeiro', ..., 12: 'Dezembro'}
meses_map = {i+1: ordem_meses[i] for i in range(12)}
df['mes'] = df['mes_numero'].map(meses_map)

# Converter a coluna 'mes' para uma categoria ordenada
df['mes'] = pd.Categorical(df['mes'], categories=ordem_meses, ordered=True)

# Remover a coluna auxiliar 'mes_numero'
df = df.drop(columns=['mes_numero'])

# Calcular a receita total por mês
receita_por_mes = df.groupby('mes')['receita'].sum().reset_index()

# Criar o gráfico de linha
plt.figure(figsize=(14, 7))
sns.lineplot(x='mes', y='receita', data=receita_por_mes, marker='o', color='green')
plt.title('Receita Total Mensal')
plt.xlabel('Mês')
plt.ylabel('Receita Total')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Distribuição da Quantidade de Itens Vendidos (`quantidade`)

Vamos agora analisar a distribuição da quantidade de itens vendidos por transação. Isso pode revelar padrões de compra, como se a maioria dos clientes compra poucos itens ou muitos em uma única transação.

```python
# Criar um histograma para a coluna 'quantidade'
plt.figure(figsize=(10, 6))
sns.histplot(df['quantidade'], bins=20, kde=True, color='skyblue')
plt.title('Distribuição da Quantidade de Itens Vendidos')
plt.xlabel('Quantidade')
plt.ylabel('Frequência')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Distribuição do Lucro Total (`lucro`)

Por fim, vamos observar a distribuição da margem de lucro. Isso é crucial para entender a saúde financeira das vendas e identificar o quão eficientemente os lucros estão sendo gerados.

### Correlação entre Quantidade e Receita

Vamos calcular a correlação entre a `quantidade` de itens vendidos e a `receita` gerada. Isso nos ajudará a entender a força e a direção da relação linear entre essas duas variáveis.

```python
# Calcular a correlação entre 'quantidade' e 'receita'
correlacao = df['quantidade'].corr(df['receita'])

print(f"A correlação entre a quantidade de itens vendidos e a receita é: {correlacao:.2f}")
```

### Lucro Total por Estado (`uf`)

Além da receita, o lucro é um indicador crucial de desempenho financeiro. Vamos visualizar o lucro total gerado por cada estado (`uf`) para complementar a análise de receita e identificar onde a operação é mais rentável.

```python
# Calcular o lucro total por estado
lucro_por_uf = df.groupby('uf')['lucro'].sum().sort_values(ascending=False).reset_index()

# Criar o gráfico de barras para lucro
plt.figure(figsize=(12, 6))
sns.barplot(x='uf', y='lucro', data=lucro_por_uf, palette='cividis')
plt.title('Lucro Total por Estado (UF)')
plt.xlabel('Estado (UF)')
plt.ylabel('Lucro Total')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Receita Total por Canal de Venda e Segmento

Para ter uma visão mais granular do desempenho, vamos analisar a receita total gerada por cada combinação de `canal_venda` e `segmento`. Este gráfico nos ajudará a entender qual canal performa melhor em qual segmento de cliente.

```python
# Calcular a receita total por canal de venda e segmento
receita_canal_segmento = df.groupby(['canal_venda', 'segmento'])['receita'].sum().reset_index()

# Criar o gráfico de barras agrupado
plt.figure(figsize=(12, 7))
sns.barplot(x='canal_venda', y='receita', hue='segmento', data=receita_canal_segmento, palette='viridis')
plt.title('Receita Total por Canal de Venda e Segmento')
plt.xlabel('Canal de Venda')
plt.ylabel('Receita Total')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Segmento')
plt.tight_layout()
plt.show()
```
