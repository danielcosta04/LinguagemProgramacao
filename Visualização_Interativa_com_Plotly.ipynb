```python
import pandas as pd
import plotly.express as px
```

First, I'll load the data from `vendas_brasil_clean_aula6_plotly.csv` into a pandas DataFrame and display the first few rows to understand its structure.

```python
df = pd.read_csv('/content/vendas_brasil_clean_aula6_plotly.csv')
display(df.head())
```

Now, let's create a simple interactive scatter plot using Plotly Express. This plot will show `total_vendas` (total sales) against `data` (date), allowing for interaction to explore trends.

```python
fig = px.line(df, x='data', y='total_vendas', title='Total de Vendas ao Longo do Tempo (Interativo)')
fig.show()
```
